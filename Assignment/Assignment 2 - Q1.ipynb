```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

cancer_data = load_breast_cancer()
X = cancer_data.data
y = cancer_data.target

print(bc_data.keys())
df = pd.DataFrame(data=X, columns=cancer_data.feature_names)

print(df.shape)

df1 = pd.DataFrame(data=y , columns=['name'])
print(df1.describe())
print(df1.head(10))
print (df1[df1['name']==1].shape)
print (df1[df1['name']==0].shape)

# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
#Build a logistic regression model on the above data

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=3000)
classifier.fit(X_train, y_train) 
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score

# predict probabilities
pred_prob = classifier.predict_proba(X_test)

# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(y_test, pred_prob[:,1], pos_label=1)

# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)

# auc scores
auc_score1 = roc_auc_score(y_test, pred_prob[:,1])

print(auc_score1)
```

```python
# matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='Logistic Regression')
#plt.plot(fpr2, tpr2, linestyle='--',color='green', label='KNN')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show();
```

```python
# based on the graph the curve is at 0.99 and should be picked as the threshold
```

```python
# predict probabilities

def custom_predict(X, threshold):
    probs = classifier.predict_proba(X) 
    return (probs[:, 1] > threshold).astype(int)
    
    
y_pred = custom_predict(X=X_test, threshold=0.9)

#pred_prob = custom_predict()
```
