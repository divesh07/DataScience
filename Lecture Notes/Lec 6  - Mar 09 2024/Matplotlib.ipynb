#### Annotations over Bar Plot

```python
'''x-coordinate: bar.get_x() + bar.get_width() / 2 
y-coordinate: bar.get_height() 
free space to be left to make graph pleasing: (0, 8) 
ha and va stand for the horizontal and vertical alignment'''

food = np.array([10, 25, 20, 15, 22])
put_labels = ["MILK", "BUTTER", "BREAD", "TEA", 'COFFEE']

x = np.arange(len(put_labels))  # the label locations
### print(x)

##width = 0.8  # the width of the bars 

plots = plt.bar(x,food)
plt.xticks(x, labels = put_labels)

#iterate over every bar
for bar in plots:
    plt.annotate(format(bar.get_height(), '.3f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=10, xytext=(0,5), 
                   textcoords='offset points')
```

#### Marking the points

```python
'''Scatter Plot'''
#hatch = to set the pattern over the bar

sample.plot(x ='Age', y = 'Salary', kind = "scatter", hatch = '//', s= 100 )

for i, txt in enumerate(sample['Name']):
    plt.annotate(txt, (sample['Age'][i], sample['Salary'][i]))
```
