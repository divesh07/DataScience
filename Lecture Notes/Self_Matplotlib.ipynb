```python
#Overlapping the plots
import numpy as np
x1 = np.array([0,1,2,3])
y1 = np.array([2,4,7,9])

x2 = np.array([0,1,2,3])
y2 = np.array([1,3,9,11])

x3 = np.array([0,1,2,3])
y3 = np.array([5,8,3,10])

plt.plot(x1,y1,x2,y2,x3,y3)
plt.title("Overlapping plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
# Bar plot and annotations
food = np.array([10,25,20,15, 22])
put_labels = ["MILK", 'BUTTER', 'BREAD', "TEA", 'COFFEE']
x = np.arange(len(put_labels))
print (x)
plots = plt.bar(x,food)
plt.xticks(x, labels=put_labels)

for bar in plots:
        plt.annotate(format(bar.get_height(), '.3f'),  
                           (bar.get_x() + bar.get_width() / 2,  
                            bar.get_height()), ha='center', va='center', 
                           size=10, xytext=(0,5), 
                           textcoords='offset points')
    
```

```python
#Marking the points

sample.plot(x='Age', y='Salary', kind= 'scatter', hatch="//", s=100)

for i,txt in enumerate(sample['Name']):
    plt.annotate(txt , (sample['Age'][i], sample['Salary'][i]))
```
