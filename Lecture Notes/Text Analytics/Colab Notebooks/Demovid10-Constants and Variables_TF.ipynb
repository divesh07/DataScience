* We don't need to use placeholders 
* We don't need to initialize variables using global_variable_initializer()

```
import tensorflow as tf
```

```
tf.executing_eagerly()
```

```
one = tf.constant(1)
```

```
tf.print(one)
```

```
ten = tf.Variable(10)
```

```
tf.print(ten) # We don't need to initialize variables using tf.global_variables_initializer()
```

# Mathematical Operations: 

* Add 2 numbers

```
sum = tf.add(one, ten)
```

```
tf.print(sum)
```

* Assign sum in ten

```
ten = sum
```

```
tf.print(ten)
```
