````
# Stopword removal - Removes helping verbs like the,a,an,is etc.
some_text1 = 'a random text is written. any random word written by us'

# corpus - collection of words in form of document 
# corpora - plural of corpus 
from nltk.corpus import stopwords
word_corpus = stopwords.words('english')
````

````
[w for w in some_text1.split() if w not in word_corpus]
````

````
# Stemming - Used for removing the affixes(root word) for eg. - going -> go
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
list_form = ['some', 'was', 'a', 'founded']
word_tokenized = [stemmer.stem(tokenization) for tokenization in list_form]
````
