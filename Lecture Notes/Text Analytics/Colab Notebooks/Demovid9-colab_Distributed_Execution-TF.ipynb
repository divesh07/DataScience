Implement a simple function -> X^2*Y+Y+C

```
import tensorflow as tf
```

```
X = 3
```

```
Y = 4
```

```
C = 2 
```

```
@tf.function
def func(a,b,c):
  result = a*a*b+b+c
  return result
```

```
value = func(X,Y,C)
```

```
tf.print(value)
```

```
value = func(10,15,2)
```

```
tf.print(value)
```
