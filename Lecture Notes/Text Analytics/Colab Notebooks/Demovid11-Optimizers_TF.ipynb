Solve the equation 25 = a*5 

```
import tensorflow as tf
```

```
a = tf.Variable(0.0)
```

```
def myloss(x):
  result = a*x
  return result
```

```
tf.print(myloss(5))
```

# Problem:
* We are getting zero coz 'a' is initialized to zero which is getting multiplied with the loss fn input value 

# Solution:
* We will make our optimizer to change the value of our 'a' to solve our equation from(25 = a*x)
* Finally it should get a value of 5 for 'a'

```
result = lambda : abs(myloss(5)-[25]) 
```

```
optimizer = tf.optimizers.Adam(1)
```

* We want to minimize the loss through loss fn 
* We want to use optimizer of type 'Adam' where (1) is the learning rate (doesn't matter very much in this case) 

```
optimizer.minimize(result, [a]) 
```

* Using minimize method and providing loss fn as one parameter
* Second parameter is a set of trainable variables which in this case is 'a' (which will be trained)

```
tf.print(a)
```

```
for x in range(1000):
  optimizer.minimize(result, [a])
```

using ranges to keep using different values for 'a' (training 'a') till the loss becomes zero

```
tf.print(a)
```
