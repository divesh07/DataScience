```python
# Seaborn - Static visualizations

import seaborn as sns
sns.get_dataset_names()
```

```python
mpg_df = sns.load_dataset('mpg')
mpg_df.head()
```

```python
#Scatter plot - represent values of 2 features in a dataset

import warnings
sns.scatterplot(x='weight', y='mpg', data= mpg_df)
warnings.filterwarnings('ignore')

# Analyze the plot
'''
As the weight increases the mpg decreases
'''
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin')
# Cars in USA are heavier and have less average
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='horsepower')
# cars that have more weight have more hp ( mostly belong to USA)
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin')
# Hue is used to represent categorical data
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='cylinders', style='origin')
#hue with changing saturation is used to represent numerical data by defult
# cars made in the usa are heavier , have less mileage and have more cylinders
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin', palette='deep')
```

```python
palette_options = ['deep','muted','pastel','bright','colorblind','dark']
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin', palette=palette_options[2])
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin', palette=palette_options[2]).set_title(palette_options[2])
```

```python
# For numerical data
palette_options = ['rocket', 'rocket_r', 'mako', 'flare', 'crest', 'magma', 'viridis']
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='cylinders', style='origin', palette=palette_options[2]).set_title(palette_options[2])
```

```python
colorPalette = sns.color_palette('Spectral')
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='cylinders', style='origin' , palette= colorPalette).set_title('Spectral')
```

```python
# Using relplot to combine scattrerplot() wtih facetgrids

sns.relplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'],col='origin')
```

```python
# adding rows
sns.relplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'],col='origin', row='cylinder')
```

```python
sns.relplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'],col='origin', height=5, aspect=0.7)
# width = aspect * height
```

```python
# Hex bin plot
# used when both rows n cols correspond to numerical attributes
sns.set(style='ticks') # to include ticks on the axis
sns.jointplot(x='weight', y='mpg', data= mpg_df) # scatter plot by default
```

```python
sns.jointplot(x='weight', y='mpg', data= mpg_df, kind = 'hex') # scatter plot by default
```

```python
sns.jointplot(x='weight', y='mpg', data= mpg_df, color = (0,0.7,0.7))  # color (r,g,b)
```

```python
# adding hue will add colors to the scatterplot
# will aslo draw density curves on the axes

sns.jointplot(x='weight', y='mpg', data= mpg_df, hue='origin') # hue and kind cannot be used together
```

```python
# Kde plot - Kernel density plot , to viz the distibution
# continuous probability density curve
sns.kdeplot(data=mpg_df)
```

```python
#scatterplot , relplot , jointplot , hexbinplot , kdeplot
sns.scatterplot(data=mpg_df, x='weight', y='mpg' , hue='origin')
sns.relplot(x='weight', y='mpg', data=mpg_df,hue ='origin', col='origin')
```

```python
sns.kdeplot(data=mpg_df, x='weight', hue='origin', fill=True)
sns.jointplot(data=mpg_df, x='weight', y='mpg', hue='origin')
sns.jointplot(data=mpg_df, x='weight', y='mpg', kind='hex')
```

```python
#Contour plots - Used when data points are densely populated ( same use case as hex bin)
# used for weather indicators

#Learnt till now - scatterplot , relplot , jointplot , hexbinplot , kdeplot
styles = ["white", "dark", "whitegrid", "darkgrid", "ticks"]
sns.set_style(styles[0])
import warnings
sns.kdeplot(data=mpg_df, x='displacement', y='weight')
warnings.filterwarnings('ignore')
```

```python
sns.kdeplot(data=mpg_df, x='displacement', y='weight', fill=True)
# Interpretation of Contour plots is similar to hexbin plots
# darker region indicate more data points
```

```python
sns.kdeplot(data=mpg_df, x='displacement', y='weight', cmap = "Reds", fill=True)
```

```python
sns.kdeplot(data=mpg_df, x='displacement', y='weight', cmap = "Reds", bw_adjust=0.5, fill=True)
```

```python
# Question : Make a scatter plot between hp and acceleration for different model year
mpg_df.head()
sns.scatterplot(data=mpg_df, x="horsepower", y='acceleration', hue='model_year')
```

```python
sns.relplot(data=mpg_df, x="horsepower", y='acceleration', hue='model_year', col='model_year')
```

```python
'''
Learnt till now - scatterplot , relplot , jointplot , hexbinplot , kdeplot ( y axis is always density), 
contour plot ( kde plot with fill , x and y axis)

line plot
 used for indicating relationship as series of data points between discrete ( data with certain values ) numerical
 feature ( on x axis ) and continuous numerical feature on y axis
'''
styles = ["white", "dark", "whitegrid", "darkgrid", "ticks"]
sns.set_style(styles[0])
sns.scatterplot(x='model_year', y='mpg', data=mpg_df)
```

```python
# model-year takes discrete values betn 70 and 82
sns.lineplot(x='model_year', y='mpg', data=mpg_df)

#Shaded areas shows confidence interval for y axis ( default = 95%) , ci param to change it
# solid line represents mean of y axis feature
```

```python
sns.lineplot(x='model_year', y='mpg', data=mpg_df, ci=30)

```
