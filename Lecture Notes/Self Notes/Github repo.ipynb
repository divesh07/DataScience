```python
'''
Make an API call to github
Paginate through repo API's
Take data and parse it
'''
```

```python
import requests
```

```python
url = "https://api.github.com/search/repositories"
url += "?q=language:python"
```

```python
headers = {"Accept": "application/vnd.github.v3+json"}
response = requests.get(url, headers= headers)
print(response.status_code)
```

```python
json_response = response.json()
json_response
```

```python
json_response.keys()
```

```python
json_response['total_count']
```

```python
response_dicts = json_response['items']
len(response_dicts)
```

```python
first_repo = response_dicts[0]
first_repo
```

```python
for key,value in first_repo.items():
    print (key)
```

```python
for response_dict in response_dicts:
    print(response_dict['name'])
    print(response_dict['stargazers_count'])
```

```python
import matplotlib.pyplot as plt
```

```python
repo_names, stars = [],[]

for response_dict in response_dicts:
    repo_names.append(response_dict['name'])
    stars.append(response_dict['stargazers_count'])
    pass
```

```python
print(repo_names)
```

```python
plt.figure(figsize=(18,6))
plt.bar(repo_names,stars)
plt.title("Github repo vs stars", color = 'green', fontsize=20)
plt.xticks(rotation = 90, color='red', fontsize=15)
plt.show()
```
