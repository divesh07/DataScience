```python
import PyPDF2
reader = PyPDF2.PdfReader('ITAct.pdf')
print(reader.metadata)
print(reader.pages)
```

```python
entiretext = ""
for i in range(0,len(reader.pages)):
    entiretext += reader.pages[i].extract_text()
```

```python
entiretext
```

```python
#?PyPDF2.PdfReader
```

```python
#n-gram dictionary
ngrams = {}
```

```python
# Building the model
n=1
words = nltk.word_tokenize(entiretext)
for i in range(len(words)-n):
    gram = ' '.join(words[i:i+n])
    if gram not in ngrams.keys():
        ngrams[gram] = []
    ngrams[gram].append(words[i+n])
```
