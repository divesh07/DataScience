```python
# Seaborn - Static visualizations

import seaborn as sns
sns.get_dataset_names()
```

```python
mpg_df = sns.load_dataset('mpg')
mpg_df.head()
```

```python
#Scatter plot - represent values of 2 features in a dataset

import warnings
sns.scatterplot(x='weight', y='mpg', data= mpg_df)
warnings.filterwarnings('ignore')

# Analyze the plot
'''
As the weight increases the mpg decreases
'''
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin')
# Cars in USA are heavier and have less average
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='horsepower')
# cars that have more weight have more hp ( mostly belong to USA)
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin')
# Hue is used to represent categorical data
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='cylinders', style='origin')
#hue with changing saturation is used to represent numerical data by defult
# cars made in the usa are heavier , have less mileage and have more cylinders
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin', palette='deep')
```

```python
palette_options = ['deep','muted','pastel','bright','colorblind','dark']
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin', palette=palette_options[2])
```

```python
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='origin', style='origin', palette=palette_options[2]).set_title(palette_options[2])
```

```python
# For numerical data
palette_options = ['rocket', 'rocket_r', 'mako', 'flare', 'crest', 'magma', 'viridis']
sns.scatterplot(x='weight', y='mpg', data= mpg_df, hue='cylinders', style='origin', palette=palette_options[2]).set_title(palette_options[2])
```

```python
colorPalette = sns.color_palette('Spectral')
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='cylinders', style='origin' , palette= colorPalette).set_title('Spectral')
```

```python
#Size parameter
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower')
```

```python
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100))
```

```python
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='full')
```

```python
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='brief')
```

```python
#Markers
sns.scatterplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'])
```

```python
# Using relplot to combine scattrerplot() wtih facetgrids

sns.relplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'],col='origin')
```

```python
# adding rows
sns.relplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'],col='origin', row='cylinder')
```

```python
sns.relplot(x='weight', y='mpg', data=mpg_df , hue='origin', style='origin', palette='pastel', size='horsepower', sizes=(20,100), legend='auto', markers=['1','2','3'],col='origin', height=5, aspect=0.7)
# width = aspect * height
```
