## Bitwise Operators:
* In Python, bitwise operators are used to performing bitwise calculations on integers. 
* The integers are first converted into binary and then operations are performed on bit by bit. 
**Finally the result is returned in decimal format.**

| Symbol | Task Performed |
|----|---|
| &  | Logical And |
| l  | Logical OR |
| ^  | XOR |
| ~  | Negate |
| >>  | Right shift |
| <<  | Left shift |

### AND (&) operator returns 1 if both the bits are 1 else 0

### Example a = 10 and b = 4
* a= 10 = 1010 (in binary)
* b = 4 = 0100 (in binary)
* 0000 (in case of & operator)
* 1110 (in case of | operator)
* ----

```python
#Example for a = 10 and b = 4
a = 10
b = 4
print(a & b)
```

```python
c=10
bin(c)
### The prefix 0b represents that a given string is a binary string
```

```python
print(a & b)
```

### Example a = 10 and b = 4 
#### To determine (a | b)
* a= 10 = 1010 (in binary)
* b = 4 = 0100 (in binary)
* 1110 (in case of | operator)

### OR (|) operator returns 1 if either of the bit is 1 else 0

```python
print(a | b)
```

### XOR (^) operator returns 1 if one of the bits is 1 and the other is 0 else returns false.

```python
print(a ^ b)
```

### not (~) operator returns one’s complement of the number.
    ones complement means adding one to the binary of the given integer
    a = 10 = 1010 (in binary)

```python
print (~a)
```

### Right Shift: Bitwise right shift: Shifts the bits of the number to the right and fills 0 on voids left

```python
a = 10 ### 0000 1010 (Binary)
a >> 1 ### 0000 0101 = 5
```

### Bitwise left shift: Shifts the bits of the number to the left and fills 0 on voids right as a result. 

```python
b=10  ### 0000 1010
b << 1 ### 0001 0100
```

## Useful built-in functions for simplifying Arithmetic Operations

**round( )** function rounds the input value to a specified number of places or to the nearest integer. 

```python
print(round(5.5231))
print(round(4.55892, 2))
```

**divmod(x,y)** outputs the quotient and the remainder in a tuple(which we will d in the next class) in the format (quotient, remainder). 

```python
divmod(9,2) ### 9/2; quotient is 4 and renmainder is 1
```

**pow(x,y,z)** can be used to find the power $x^y$ also the mod of the resulting value with the third specified number can be found i.e. : 
## ($x^y$ % z).

```python
print(pow(3,3)) ### 3^3
print(pow(3,3,5)) ### 3^3=27; 27%5 = 2
```
