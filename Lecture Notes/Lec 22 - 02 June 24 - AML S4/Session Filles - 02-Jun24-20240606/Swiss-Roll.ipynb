```python
# Importing Libraries:
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
from sklearn import manifold, datasets
from sklearn.decomposition import PCA
```

```python
#Creating Swiss-Roll Data:
X, y = datasets.make_swiss_roll(800)
# Creating the plot
fig = plt.figure(figsize=(20, 20))
# Adding 3d scatter plot
ax = fig.add_subplot(231, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Spectral)
ax.view_init(4, -72)
```

```python
#Applying PCA:
pca = PCA(n_components=2) # creates an instance of PCA class
results = pca.fit(X) # applies PCA on predictor variables
Z = results.transform(X) # create a new array of latent variables
plt.scatter(Z[:,0],Z[:,1], c=y, cmap=plt.cm.Spectral)
```

```python
#Applying MDS:
from sklearn.manifold import MDS
mds = MDS(n_components=2)
X_2d = mds.fit_transform(X)
plt.scatter(X_2d[:,0],X_2d[:,1], c=y, cmap=plt.cm.Spectral)
```

```python
#Applying ISOMAP:
from sklearn.manifold import Isomap
model = Isomap(n_components=2)
X_2d_i = model.fit_transform(X)

plt.scatter(X_2d_i[:,0],X_2d_i[:,1], c=y, cmap=plt.cm.Spectral)
```

```python
#Applying LLE:
from sklearn.manifold import LocallyLinearEmbedding
X_lle = LocallyLinearEmbedding(n_neighbors=10, n_components=2).fit_transform(X)

plt.scatter(X_lle[:,0],X_lle[:,1], c=y, cmap=plt.cm.Spectral)
```

```python
#Applying t-SNE:
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2).fit_transform(X)

plt.scatter(X_tsne[:,0],X_tsne[:,1], c=y, cmap=plt.cm.Spectral)
```
