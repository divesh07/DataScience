```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox
import sklearn
from sklearn import (manifold, datasets, decomposition, ensemble,
                     discriminant_analysis)
```

```python
digits = datasets.load_digits(n_class=10)
X = digits.data
y = digits.target
n_samples, n_features = X.shape
n_neighbors = 30
X.shape
```

```python
# Scale and visualize the embedding vectors
def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure()
    ax = plt.subplot(111)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(y[i]),
                 color=plt.cm.Set1(y[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    
    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)

```

```python
#PCA:
# Projection on to the first 2 principal components
from sklearn.decomposition import PCA
X_pca = PCA(n_components=2).fit_transform(X)
```

```python
#LLE
from sklearn.manifold import LocallyLinearEmbedding
X_lle = LocallyLinearEmbedding(n_neighbors=70, n_components=2).fit_transform(X)
```

```python
#isomap
from sklearn.manifold import Isomap
X_is = Isomap(n_components=2).fit_transform(X)
```

```python
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2).fit_transform(X)
```

```python
#Isomap:
plot_embedding(X_pca,
               "Principal Components projection of the digits (time %.2fs)" )

plot_embedding(X_is,
               "Isomap Visualization of the digits" )

plot_embedding(X_lle,
               "LLE Visualization of the digits" )

plot_embedding(X_tsne,
               "t-SNE Visualization of the digits" )
```
