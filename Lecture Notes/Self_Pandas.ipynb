```python
#Data Frames
# Prepare Data
# Define rows and cols
Data = np.array([[1,1,1],[2,4,8],[3,9,27],
                 [4,16,64],[5,25,125],[6,36,216],
                 [7,49,343]])
print(Data)

# Define rows
Rows = ['first','second','third','fourth','fifth','sixth','seventh']
Cols = ['1','squares','cubes']
df_withoutLabel = pd.DataFrame(data=Data) #Data without labels
df = pd.DataFrame(data=Data, index=Rows, columns=Cols)
print(df_withoutLabel)
print(df)
```

```python
# Adding a new row into DF

df_1 = pd.DataFrame({"A":[1,2,3], "B":[5,6,7]})
df_2 = pd.DataFrame({"A":[10,20,30,40], "B":[50,60,70,80]})
df_1
```

```python
df_2
```

```python
# use concat or _append
df_3 = df_1._append(df_2)
df_3
```

```python
df_3 = df_1._append(df_2,ignore_index=True)
df_3
```

```python
df_4 = df_3.drop(3,axis=0)
df_4
```

```python
df_4.reset_index(inplace = True, drop = True)
df_4
```

```python
# appending df with diff shapes
df_5 = pd.DataFrame({"A":[1,2,3,4], "B":[5,6,7,8]})
print(df_5)
df_6 = pd.DataFrame({"A":[10,20,30],
                     "B":[50,60,70],
                     "C":[80,90,100]})
print(df_6)
```

```python
df_7 = df_5._append(df_6, ignore_index=True)
df_7
```

```python
# Dealing with missing values
df_7.isna()
```

```python
# Count no of missing values
df_7.isna().sum()
```

```python
# fill with mean of other values
df_7.fillna(df_7["C"].mean(), inplace=False) # inplace false would not make changes to org df
df_7
```

```python
df_7.fillna(df_7["C"].mean(), inplace=True) 
df_7
```

```python
#drop missing value col
df_7.dropna() # by default , inplace = false
df_7
```
